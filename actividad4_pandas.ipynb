```python
# ============================================================
# Actividad 4: Manejo de Datos con Pandas
# ============================================================

import pandas as pd
import numpy as np

# ============================================================
# Dataset base
# ============================================================

datos = {
    'nombre':   ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'edad':     [25, 30, 22, None, 28],
    'ciudad':   ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'producto': ['Laptop', 'Teléfono', 'Tablet', 'Laptop', 'Tablet'],
    'precio':   [1200, 800, 300, 1150, None],
    'stock':    [10, 15, 5, 8, 0]
}

df = pd.DataFrame(datos, index=['a', 'b', 'c', 'd', 'e'])
df.to_csv('actividad_semana4.csv', index=True)
print("=== DataFrame Base ===")
display(df)

# ============================================================
# 1) SERIES — crear y operar
# ============================================================

serie_lista = pd.Series([10, 20, 30, 40], index=['x', 'y', 'z', 'w'])
print("\n=== Serie desde lista ===")
display(serie_lista)

serie_dic = pd.Series({'manzana': 5, 'banana': 8, 'naranja': 3})
print("\n=== Serie desde diccionario ===")
display(serie_dic)

print("\nValor de 'banana':", serie_dic['banana'])
serie_dic['manzana'] = 10
print("Serie modificada ('manzana' = 10):")
display(serie_dic)

print("Serie multiplicada por 2:")
display(serie_dic * 2)

# ============================================================
# 2) DATAFRAME — crear y explorar
# ============================================================

datos_df = {
    'nombre': ['Luis', 'María', 'José'],
    'edad': [21, 35, 27],
    'ciudad': ['Quito', 'Caracas', 'Bogotá']
}
df2 = pd.DataFrame(datos_df, index=['x', 'y', 'z'])
print("\n=== DataFrame creado manualmente ===")
display(df2)

print("\nColumna 'nombre':")
display(df2['nombre'])

print("\nFila con loc['x']:")
display(df2.loc['x'])

print("\nFila con iloc[0]:")
display(df2.iloc[0])

# ============================================================
# 3) OPERACIONES BÁSICAS
# ============================================================

df['precio_descuento'] = df['precio'] * 0.9
df['stock'] = df['stock'] + 5
print("\n=== DataFrame con columna derivada y stock actualizado ===")
display(df)

# ============================================================
# 4) MANEJO DE DATOS FALTANTES
# ============================================================

print("\n=== Detección de nulos ===")
display(df.isna())

print("\n=== Conteo de faltantes por columna ===")
display(df.isna().sum())

df['edad'] = df['edad'].fillna(0)
df['ciudad'] = df['ciudad'].fillna('Desconocido')
df['precio'] = df['precio'].fillna(0)
print("\n=== DataFrame sin valores nulos ===")
display(df)

# ============================================================
# 5) SELECCIÓN Y FILTRADO
# ============================================================

print("\n=== Filas con precio > 500 ===")
display(df[df['precio'] > 500])

print("\n=== Productos 'Laptop' con stock > 5 ===")
display(df[(df['producto'] == 'Laptop') & (df['stock'] > 5)])

# ============================================================
# 6) ORDENAR DATOS
# ============================================================

print("\n=== Ordenado por edad (ascendente) ===")
display(df.sort_values('edad'))

print("\n=== Ordenado por precio (descendente, llenando nulos con 0) ===")
display(df.sort_values('precio', ascending=False))

# ============================================================
# 7) ESTADÍSTICAS BÁSICAS
# ============================================================

print("\n=== Estadísticas de columnas numéricas ===")
display(df.describe())

print("\n=== Conteo de valores en 'producto' ===")
display(df['producto'].value_counts())

# ============================================================
# 8) LEER Y GUARDAR DATOS
# ============================================================

df_leido = pd.read_csv('actividad_semana4.csv')
print("\n=== CSV leído desde archivo ===")
display(df_leido.head())

df[['nombre', 'producto', 'precio_descuento']].to_csv('datos_filtrados.csv', index=False)
print("\nArchivo 'datos_filtrados.csv' guardado correctamente.")

# ============================================================
# 9) EJERCICIO INTEGRADO
# ============================================================

df['precio_descuento'] = df['precio'] * 0.9
df_filtrado = df[df['stock'] > 5]
df_ordenado = df_filtrado.sort_values('precio_descuento')
df_ordenado.to_csv('inventario_procesado.csv', index=False)

print("\n=== DataFrame final procesado ===")
display(df_ordenado)
print("\nArchivo 'inventario_procesado.csv' generado exitosamente.")
```
